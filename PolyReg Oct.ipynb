```python
import numpy as np
from pylab import *
import matplotlib.pyplot as plt

height = np.random.normal(10,2,1000)
weight = np.random.normal(60,30,1000)/(height*0.5)

scatter(height,weight);
```

```python
xp = np.poly1d(np.polyfit(height,weight,6)) 
```

```python
plt.scatter(height,weight)
plt.plot(height,xp(height),c='r')
```
